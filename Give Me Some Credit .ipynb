```python
train = pd.read_csv('/kaggle/input/GiveMeSomeCredit/cs-training.csv')
train.head()
```

```python
test = pd.read_csv('/kaggle/input/GiveMeSomeCredit/cs-test.csv')
test.head()
```

```python
y = train['SeriousDlqin2yrs']
y
```

```python
train.dtypes
```

```python
x_train = train.drop(['Unnamed: 0','SeriousDlqin2yrs'],1)
x_train.head()
```

```python
x_test = test.drop(['Unnamed: 0','SeriousDlqin2yrs'],1)
x_test.head()
```

```python
train.isnull().sum()
```

```python
from lightgbm import LGBMClassifier
```

```python
lgb = LGBMClassifier(nthread=4, colsample_bytree = 0.7, num_leaves= 5)
lgb.fit(x_train, y)
```

```python
result = lgb.predict_proba(x_test)
```
